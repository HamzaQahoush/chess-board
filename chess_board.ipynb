```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class ChessBoard () :
    blue = [0,1,1]
    red = [1,.2,0]
    white = [1,1,1]
    black = [0,0,0]

    def __init__ (self) : 
        self.grid= np.ones((8,8,3))
        
        for i in range(8): 
            for j in range (8):
                if j%2 == 0 and i%2 == 1:
                    self.grid[i,j] = ChessBoard.black
                elif j%2 == 1 and i%2 == 0:
                    self.grid[i,j] = ChessBoard.black    
        
    
    def add_red(self,row,columns):
        self.red_x=row
        self.red_y=columns
        self.grid[row,columns]=ChessBoard.red
    def add_blue(self,row,columns):
        self.blue_x=row
        self.blue_y=columns
        self.grid[row,columns]=ChessBoard.blue
    def render (self): 
        plt.imshow(self.grid)

    def is_under_attack(self,): 
        if self.red_x == self.blue_x or  self.red_y==self.blue_y :return True  
        elif abs(self.red_x - self.blue_x) == abs (self.red_y - self.blue_y ) : return True
        else : return False 
    
chess= ChessBoard()
chess.render()
# chess.add_red(3,4)
# chess.add_blue(3,5)
# chess.add_blue(4,3)
# chess.add_red(0,0)
# chess.add_blue(7,7)
chess.add_red(0,3)
chess.add_blue(1,2)
chess.is_under_attack()
chess.render()
```

```python
def test_chess_row(): 
    chess=ChessBoard()
    chess.add_red(6,5)
    chess.add_blue(6,1)
    expected= True
    actual = chess.is_under_attack()
    print(actual)
    chess.render()
    assert actual == expected 
    print ('under attack')

test_chess_row()
```

```python
def test_chess_column(): 
    chess=ChessBoard()
    chess.add_red(1,6)
    chess.add_blue(3,6)
    expected= True
    actual = chess.is_under_attack()
    print(actual)
    chess.render()
    assert actual == expected 
    print ('under attack')

test_chess_column()
```

```python
def test_chess_diagonal(): 
    chess=ChessBoard()
    chess.add_red(7,0)
    chess.add_blue(0,7)
    expected= True
    actual = chess.is_under_attack()
    print(actual)
    chess.render()
    assert actual == expected 
    print ('under attack')

test_chess_diagonal()
```
